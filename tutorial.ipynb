# Taxodist Tutorial

This notebook serves as a tutorial to show you the basic use of our package.

In our example we are going to calculate and plot the distances of a few "random" ICD-sets.

````python
# import needed libraries
from src.taxodist import td_utils as utils
from src.taxodist import tree_parsers
from src.taxodist import td_calc


tree = tree_parsers.getICD10GMTree(version='2021')
td = td_calc.Taxodist()
````

Executing the cell aboves loads the 2021 version of the ICD-10-GM taxonomy to our ```tree``` variable and the calculation-instance into ```td```.

Next we are going to generate some test data.

````python
sets = [utils.getRandomConcepts(15,tree) for x in range(0,10)]

matrix = td.calc_set_sim(sets, tree,'levels','path_based','mean_cs')
````

A call to `td.calc_set_sim()` allows you to choose the algorithm-combination. You could also use `utils.getIC()` or `utils.getCS()` to tetrieve the single values of a certain concept or concept concept-pair.

`matrix` holds the pair-wise similarity matrix of the random sets.
Let us take a look t the results by creating a heatmap.

````python
import pandas as pd

df = pd.DataFrame(matrix)
df.style.background_gradient(cmap='Blues')
````
